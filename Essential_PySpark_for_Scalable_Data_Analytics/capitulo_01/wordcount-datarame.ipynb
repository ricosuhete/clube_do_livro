```
from pyspark.sql.functions import split, explode

linesDf = spark.read.text("/databricks-datasets/README.md")
wordListDf = linesDf.select(split("value", " ").alias("words"))
wordsDf = wordListDf.select(explode("words").alias("word")) # explode -> Cria uma coluna com os valores que sofreram o split
wordCountDf = wordsDf.groupBy("word").count() # Conta as palavras únicas
wordCountDf.show()
wordCountDf.write.csv("/tmp/wordcounts.csv")
```
