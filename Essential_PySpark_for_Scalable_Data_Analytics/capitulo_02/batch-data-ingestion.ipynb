```
spark.conf.set("spark.sql.adaptive.enabled", "true")
```

```
retail_df = (spark
                 .read
                 .option("header", "true")
                 .option("inferSchema", "true")
                 .csv("/FileStore/tables/online_retail/online_retail.csv")
            )
```

```
retail_df.rdd.getNumPartitions()
```

```
(retail_df
       .write
       .mode("overwrite")
       .parquet("/tmp/data-lake/online_retail.parquet")
)
```
